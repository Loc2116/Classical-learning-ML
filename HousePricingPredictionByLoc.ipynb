```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
import pandas as pd
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#loading the dataset
HousePrice = pd.read_csv('HousePricingPrediction.csv', header=0)
```

```python
HousePrice.head()
```

```python
#number of rows and columns
HousePrice.shape
```

```python
HousePrice.describe()
```

```python
print(HousePrice.columns)
```

```python
HousePrice = HousePrice[HousePrice['price'] > 0]
```

```python
#seperate test
X= HousePrice.drop(['price'], axis = 1)
Y = HousePrice['price']
```

```python
#cleaning
X = X.drop(['date', 'street', 'country'], axis=1)
print(X)
print(Y)
```

```python
le= LabelEncoder()
X['city'] = le.fit_transform(X['city'])
X['statezip'] = le.fit_transform(X['statezip'])
```

```python
#Data split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Model

Thuật toán Hồi quy Tuyến tính (Linear Regression)


```python
model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5)
model.fit(X_train_scaled, Y_train)
```

Model Evaluation

```python
X_train_prediction = model.predict(X_train_scaled)
X_train_scaled_mse = metrics.mean_squared_error(Y_train, X_train_prediction)
```

```python
print('MSE score of the trainning', X_train_scaled_mse)
```

```python
X_test_prediction = model.predict(X_test_scaled)
X_test_scaled_mse = metrics.mean_squared_error(Y_test, X_test_prediction)
print('MSE score of trainning', X_test_scaled_mse)
```
